### 1. Importing libraries and files

```python
import pandas as pd
import numpy as np
import networkx as nx
import os
import matplotlib.pyplot as plt
from pyvis.network import Network
import seaborn as sns
import pyvis

relationship_df = pd.read_csv('relationships.csv')
```

### 2. Creating visualizations

```python
# Create a graph from a pandas dataframe

G = nx.from_pandas_edgelist(
    relationship_df, source = "source", target = "target", edge_attr =  "value", create_using = nx.Graph()
)
```

```python
# Create a Kamada Kawai Layout
plt.figure(figsize = (10,10))
pos = nx.kamada_kawai_layout (G)
nx.draw(G, with_labels = True, node_color = 'red', edge_cmap = plt.cm.Blues, pos = pos)
```

```python
# Define net object

net = Network()
net = Network(notebook = True,width="900px", height="800px", bgcolor='#222222', font_color='white')

# Define nodes from the G object
node_degree = dict(G.degree)

# Setting up node size attributes
nx.set_node_attributes(G, node_degree, 'size')
net.from_nx(G)
net.show_buttons(filter_= True)
net.repulsion()
net.show("country.html")
```

```python
# Apply the Leiden algorithm for community detection

from cdlib import algorithms
coms = algorithms.leiden(G)
```

```python
new_com = coms.to_node_community_map()
new_com
```

```python
# Put the output from new_com into a dictionary

dict_com = {k:v[0] for k, v in new_com.items()}
```

```python
# Rerun network graph with the communities as an argument

nx.set_node_attributes(G, dict_com, 'group')
com_net = Network(notebook = True, width="1300px", height="700px", bgcolor='#222222', font_color='white')
com_net.from_nx(G)
com_net.show_buttons(filter_ = True)
com_net.repulsion()
com_net.show("country_communities_leiden.html")
```

### Visualization Analysis
In this network visualization, we can clearly see that Germany is somewhat at the center of this network. In addition, most of its immediate edges with other countries are thick, meaning that Germany appears in conjunction with France or with Russia. Since WW1 and WW2 were some of the biggest events in the history of the 20th century, it makes sense that countries that participated in these wars have a higher significance in the chart.

We can also notice that geographically neighboring countries are often neighbors in the chart. This also makes sense since neighboring countries tend to interact with each other more often in treaties, armed conflicts, or simple negotiations. A couple of examples are the Baltic countries in blue, which are very connected, or Asian countries in red.

Japan may be the second most important node in this network, and this is not only shown by its centrality but also by the thickness of the edges. Most of its connections are with Asian countries, but its strongest connection is with Germany. They were allies in WW2, which could explain this visualization.

However, I found some unexpected relationships, such as the bottom pink network. It is clear that it mostly includes Middle Eastern countries, but it also includes Belarus and Ukraine.

All in all, with some knowledge of 20th-century history, this chart is very understandable. However, some unknown connections are worthy of further study.

### 3. Centrality Analysis

3.1 Degree Centrality

```python
# Examine degree centrality

degree_dict = nx.degree_centrality(G)


# Convert the dictionary to a DataFrame
df_degree = pd.DataFrame(list(degree_dict.items()), columns=['Country', 'Degree Centrality'])
df_degree = df_degree.sort_values(by='Degree Centrality', ascending=False)
# Choose the top 10 from the dataframe
df_degree = df_degree.head(10)
df_degree
```

```python
# Plotting degree centrality values
sns.barplot(x='Degree Centrality', y='Country', data=df_degree, orient='h')

# Add labels and title
plt.xlabel('Degree Centrality')
plt.ylabel('')
plt.title('Top 10 Countries with the Highest Degree Centrality')


# Show the plot
plt.show()
```

This chart shows that Germany and Japan have the highest degree of centrality. This matches with its placement in the network above. Seven out of the top ten countries are from Europe, highlighting Europe’s importance in this network. 

3.2 Closeness Centrality

```python
# Examine closeness centrality

closeness_dict = nx.closeness_centrality(G)


# Convert the dictionary to a DataFrame
df_closeness = pd.DataFrame(list(closeness_dict.items()), columns=['Country', 'Closeness Centrality'])
df_closeness = df_closeness.sort_values(by='Closeness Centrality', ascending=False)
# Choose the top 10 from the dataframe
df_closeness = df_closeness.head(10)
df_closeness
```

```python
# Plotting closeness centrality values
sns.barplot(x='Closeness Centrality', y='Country', data=df_closeness, orient='h')

# Add labels and title
plt.xlabel('Closeness Centrality')
plt.ylabel('')
plt.title('Top 10 Countries with the Highest Closeness Centrality')


# Show the plot
plt.show()
```

Closeness centrality reflects how efficiently a country connects with others in a network, with high values indicating strong influence and direct connections. Germany's role in global conflicts and European rebuilding made it a central hub, while Japan's economic growth and regional importance ensured strong global links. France, with its political and cultural influence and close ties in Europe, stayed well-connected. These countries' high closeness centrality shows their major role in global interactions during the 20th century.

3.3 Betwenness Centrality

```python
# Examine betweenness centrality

betweenness_dict = nx.betweenness_centrality(G)


# Convert the dictionary to a DataFrame
df_betweenness = pd.DataFrame(list(betweenness_dict.items()), columns=['Country', 'Betweenness Centrality'])
df_betweenness = df_betweenness.sort_values(by='Betweenness Centrality', ascending=False)
# Choose the top 10 from the dataframe
df_betweenness = df_betweenness.head(10)
df_betweenness
```

```python
# Plotting betweenness centrality values
sns.barplot(x='Betweenness Centrality', y='Country', data=df_betweenness, orient='h')

# Add labels and title
plt.xlabel('Betweenness Centrality')
plt.ylabel('')
plt.title('Top 10 Countries with the Highest Betweenness Centrality')


# Show the plot
plt.show()
```

Betweenness centrality shows how often a country acts as a bridge connecting others in a network, which indicates its role in facilitating connections. Germany served as a major link in European and global interactions, while Japan acted as a connector in Asia-Pacific relations and international trade. Russia, positioned between Europe and Asia, played a critical role in bridging multiple regions. These countries’ high betweenness centrality highlights their importance in linking global networks during the 20th century.
